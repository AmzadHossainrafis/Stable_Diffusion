```python
# read all the rows and add the values to the list with out the image id
rows = df.values.tolist()
print(rows[0])
print(rows[0][1:])
```

```python
#convert all the -1 to 0 
for row in rows:
    for i in range(1, len(row)):
        if row[i] == -1:
            row[i] = 0

print(rows[0])
```

```python
import pandas as pd 

dir_image = r"/home/amzad/Downloads/flicker/dataset/Flicker8k_Dataset/"
dir_text = r'/home/amzad/Desktop/stable_diffusion/dataset/Flickr8k.token.txt'



with open(dir_text, 'r') as file:
    lines = file.readlines() 

lines[0]

```

```python
#split the lines into image id and the text 
image_id = []
text = []
import os    

for line in lines:
    parts = line.split()
    join_image_path = os.path.join(dir_image, parts[0].split('#')[0])
    image_id.append(join_image_path)
    text.append(' '.join(parts[1:]))

print(image_id[0])
print(text[0])

```

```python
from PIL import Image

img = Image.open(image_id[1]) 
img.show()
```

```python
list_dir = os.listdir(dir_image)
```

```python
# make a data frame with the image id and the text
df = pd.DataFrame(list(zip(image_id, text)), columns = ['image_id', 'text'])
df.head()
reset_index = df.reset_index()
df.to_csv('flicker.csv', index = False)
```
